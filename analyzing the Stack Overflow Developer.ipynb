Task 2

```python
questions = df1['qname'].tolist()
questions
```

```python
valid_columns = list(set(df.columns) & set(questions))
df_questions = df[valid_columns]
all_answ=df_questions.dropna()
count_all_answ=all_answ.shape[0]

print('Кількість респондентів, які відповіли на всі питання: ',count_all_answ)
```

Task 3

```python
work_exp= df['WorkExp'].dropna()

mean_value=work_exp.mean()
median_value=work_exp.median()
mode_value=work_exp.mode()[0]

values_result=pd.Series({
    'Середнє:': mean_value,
    'Медіана:': median_value,
    'Мода:': mode_value,
        })

print(values_result)
```

```python
remote_counts = df['RemoteWork'].value_counts()['Remote']
print('\nКількість працівників що працюють віддалено :')
print(remote_counts)
```

Task 5

```python
worked_with_python =df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)
python_count=worked_with_python.sum()
python_percent=(python_count/total_respondents)*100
print(f'Відсоток респондентів, які програмують на Python: {python_percent:.1f}%')
```

Task 7

```python
df_worked_with_python = df[df['LanguageHaveWorkedWith'].str.contains('python', case=False, na=False)]
df_worked_with_python = df_worked_with_python.dropna(subset=['ConvertedCompYearly'])
agg_stats=(
    df_worked_with_python.groupby('Country')['ConvertedCompYearly']
    .agg(['mean','median']).round()
    .reset_index()
)
pd.set_option('display.max_rows',None)
agg_stats
```
